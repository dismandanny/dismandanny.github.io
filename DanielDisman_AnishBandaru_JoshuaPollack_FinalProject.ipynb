Which the various issues with the dataframe listed, we now can begin cleaning and transforming the data to fit the needs of our analysis later on

```python
# The following function will remove the dollar signs, commas, extra whitespace, and convert the numbers to floats
def clean_and_convert_to_float(x):
    # First check if x is not null
    if pd.notnull(x):
        # Remove the dollar sign, remove commas, remove trailing spaces, and convert to float
        return float(x.replace('$', '').replace(',', '').strip())
    else:
        # Return None if the input is null
        return None

# Apply the function to the price and service fee columns
df['price'] = df['price'].apply(clean_and_convert_to_float)
df['service fee'] = df['service fee'].apply(clean_and_convert_to_float)
df[['price', 'service fee']]
```

# **Hypothesis Testing**

**ANOVA Test**

**Now we will create a new hypothesis test on "whether the neighbourhood group of an airbnb location in New York City has a correlation with the price of the airbnb location"**

For our hypothesis testing, we will assume that the alpha value is 0.05

We will be conducting an Anova test with the following Null Hypothesis and Alternative Hypothesis:

*H0*: The neighbourhood group of an airbnb location in New York City does not have an effect on the price of the airbnb location

*HA*: The neighbourhood group of an airbnb location in New York City does have an effect on the price of the airbnb location

Anova tests are used when we are comparing the means of multiple groups, particularly three or more groups to find out if there are significant differences among them. So, in this scenario, we have five groups being the five New York City boroughs (or neighbourhood groups) and we are comparing the mean prices for each of the groups to see if there is a notable difference. If there is a big enough difference then we can conclude that there is a statistically significant correlation between the neighbourhood group of an airbnb location in New York City and the price. If there is not a big enough difference then we can conclude that there is no correlation between the neighbourhood group of an airbnb location in New York City and the price.

```python
# Checking the different neighbourhood groups available in the data set
df['neighbourhood group'].value_counts()
```

```python
# We need to fix our data as there are two values of brookln and manhatan,
# which are incorrect spellings of Brooklyn and Manhattan.
df.loc[df['neighbourhood group']=='brookln', 'neighbourhood group'] = 'Brooklyn'
df.loc[df['neighbourhood group']=='manhatan', 'neighbourhood group'] = 'Manhattan'
df['neighbourhood group'].value_counts()
```

```python
# Now we're going to check for null values in the "neighbourhood group" column and
# if there are null values, we will drop those rows because there isn't really a
# reliable way to impute this categorical data without having unwanted effects

print(df['neighbourhood group'].isnull().value_counts())
df = df.dropna(subset=['neighbourhood group'])
df['neighbourhood group'].isnull().value_counts()
# Data that is null is less than 1% of the total rows, thus dropna would work in
# this situation
```

```python
# Now we're going to check for null values in the "price" column and
# if there are null values, we will check to see if the data is skewed
print(df['price'].isnull().value_counts())
df.groupby('neighbourhood group')['price'].apply(pd.Series.skew)
```

```python
# Since there are null values and the prices for each neighbourhood group is not
# very skewed, we will use the mean price of each neighbourhood group for imputation

df['price'] = df.groupby('neighbourhood group')['price'].transform(lambda x: x.fillna(x.mean()))

# Now we will check to make sure that the code worked and that there are not more null price values:
print(df['price'].isnull().value_counts())
```

```python
# We will conduct an Anova test and find out the p-value
from scipy.stats import f_oneway

Bronx_prices = df[df['neighbourhood group'] == 'Bronx']['price']
Brooklyn_prices = df[df['neighbourhood group'] == 'Brooklyn']['price']
Manhattan_prices = df[df['neighbourhood group'] == 'Manhattan']['price']
Queens_prices = df[df['neighbourhood group'] == 'Queens']['price']
Staten_island_prices = df[df['neighbourhood group'] == 'Staten Island']['price']

f_statistic, p_value = f_oneway(Bronx_prices, Brooklyn_prices, Manhattan_prices, Queens_prices, Staten_island_prices)

print("P-value:", p_value)
```

```python
# Here we are displaying a column chart to also visualize the different mean
# prices for each neighbourhood group
import matplotlib.pyplot as plt

mean_prices = df.groupby('neighbourhood group')['price'].mean()

mean_prices.plot(kind='bar', figsize=(10, 6), color='skyblue')

plt.title('Mean New York City Airbnb Prices by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Mean Price')
plt.xticks(rotation=45)
plt.ylim(500, mean_prices.max() + 10)

plt.show()
```

**Conclusion**:
From our Anova test, we have found a p-value of 0.1387717863234742. We have assumed Î± (alpha) is 0.05. Since p-value > Î±, we fail to reject the null hypothesis. Therefore, we are accepting the null hypothesis to be true, meaning that we can assume that the neighbourhood group of an airbnb location in New York City does not have an effect on the price of the airbnb location.

**Chi Squared Hypothesis Test**



**Now we will create another hypothesis test on "Whether there is an association between the neighbourhood group of an air bnb location and the likelihood of a host being verified or not"**

For our hypothesis testing, we will assume that the alpha value is 0.05

We will be conducting an Chi squared test with the following Null Hypothesis and Alternative Hypothesis:

H0: The neighbourhood group of an airbnb location in New York City does not have an effect on the likelihood of a host being verified or not

HA: The neighbourhood group of an airbnb location in New York City does have an effect on the likelihood of a host being verified or not

```python
# Since the null values in the neighbourhood group column have been filtered and
# dropped in the previous test, we only need to filter the 'host_identity_verified'
# column.

print(df['host_identity_verified'].isnull().value_counts())
```

```python
# As we can observe, there are only 265 null rows with a null value for the
# 'host_identity_verified' column. This is less than 1% of all rows so we can simply
# perform a dropna operation to remove these rows.

df = df.dropna(subset=['host_identity_verified'])
df['host_identity_verified'].isnull().value_counts()

# We observe that we have successfully removed the null values, and the operation worked
```

```python
# Now we can finally begin the statistical hypothesis testing portion.
# We will create a 2x2 contigency table mapping New York neighbourhood to the
# verification status of the hosts
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['neighbourhood group'], df['host_identity_verified'])
contingency_table
# The contigency table is shown below ->
```

```python
# Now we will find the plot the contigency table
contingency_table.plot(kind='bar', color=['blue', 'red'], edgecolor='black')
plt.title('Number of Verified Hosts Per New York City Borough')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Hosts Verified')
```

```python
# We will now conduct the chi squared test and extract the p-value

p_value = chi2_contingency(contingency_table)[1]
p_value

# p value is shown below ->
```

**Conclusion:** From the Chi-Squared Test conducted, we observe the p-value of 0.5773674786721417 which is greater than the assumed alpha value of 0.05, thus we fail to reject the null hypothesis. This is because there is not enough evidence to suggest that there is a statistically significant relationship between the New York City neighbourhood group and whether or not an air bnb host is verified or unconfirmed. Therefore we accept that the neighbourhood group of an airbnb location in New York City does not have an effect on the likelihood of a host being verified or not. In the larger scope of the dataset and of air bnb listing in New York City, the likelihood of a host being verified would not be more apparent in one neighbourhood group over another.

# **Primary Analysis and Machine Learning**

We are choosing supervised machine learning for our primary analysis. Because we want to determine which features have the most significance on whether an Airbnb will be rated well (> 3.0 stars), classification to determine the correct projected rating seems most appropriate in this scenario. We will need to determine which model will work best for our dataset and has the best accuracy among other scores such as recall, precision, and F1. The following machine learning section will go through the process of separating the data into training and testing data, one hot encoding the categorical variables, and using the stratified k fold technique for each model to validate and observe scores for Random Forest, Decision Tree, and Logistic Regression models.

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

X = df[['host_identity_verified', 'neighbourhood group', 'instant_bookable', 'cancellation_policy', 'room type']]
y = (df['review rate number'] > 3.0)

# Clean data by removing rows with any NaN values and aligning the y with the removed rows from X
mask = X.notna().all(axis=1)
X_clean = X[mask]
y_clean = y[mask]

# Splitting data into training and testing sets
seed = 42
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.3, random_state=seed)

# Encoding categorical variables with OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')  # Add handle_unknown='ignore'
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)

# Define models
models = {
    'DecisionTree': DecisionTreeClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier()
}

# Perform k-fold (5 fold) Cross-Validation for each model
k_folds = 5
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

for model_name, model in models.items():
    cross_val = cross_val_score(model, X_train_encoded, y_train, cv=skf)
    print(f"Model: {model_name}")
    print(f"Cross validation mean: {cross_val.mean()}")
    print(f"Cross validation accuracy std dev: {cross_val.std()}\n")
```

As we observed, all cross validation means were nearly identical, with Decision Tree and Random Forest having similar standard deviations and Logistic Regression having a very small standard deviation which is good for cross validation.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(f"Number of features in X_train_encoded: {X_train_encoded.shape[1]}")
print(f"Number of features in X_test_encoded: {X_test_encoded.shape[1]}")
# Train each model using the training data
for model_name, model in models.items():
    model.fit(X_train_encoded, y_train)

# Evaluate the performance of each model
for model_name, model in models.items():
    y_pred = model.predict(X_test_encoded)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {model_name}: {accuracy}")
    print(f"Classification Report for {model_name}:")
    print(classification_report(y_test, y_pred))
```

**Why we chose the Random Forest model**: This data revealed to us that the training models had very similar results. In the end, we went with Random Forest because its only difference from Decision Tree was its slightly higher accuracy. Additionally, throughout our homework we have seen that Random Forest has a tendency to be a great model and very reliable. Logistic Regression seemed to have warnings and a True recall, precision, and F1 score of 0.0, deeming that model not good enough. With all factors considered, Random Forest was the best method for training and testing the data in addition to easily gathering various feature importances learned from the model itself.
